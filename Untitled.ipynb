```python
import csv
import functools
import os
import re
import typing

import contextily
import geopandas
import pandas
import requests
import shapely.geometry
import tqdm.auto as tqdm
from matplotlib import pyplot

T_DataFrame = pandas.core.frame.DataFrame
tqdm.tqdm.pandas()
```

# 1. Download latest NYC 311 data from NYC OpenData

**NOTE**: this is a very large file - 13.2GiB as of 2020-07-23. Extra documentation concerning field descriptors and other ways of interacting with the data can be found on [the official page](https://nycopendata.socrata.com/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9).

```python
NYC_311_URL = "https://nycopendata.socrata.com/api/views/erm2-nwe9/rows.csv?accessType=DOWNLOAD"
NYC_311_RAW_DATA = "nyc311.csv"
NYC_311_TRIMMED_DATA = "nyc311-fireworks.csv"

NYC_311_DOCS_URL = "https://nycopendata.socrata.com/api/views/erm2-nwe9/files/68b25fbb-9d30-486a-a571-7115f54911cd?download=true&filename=311_SR_Data_Dictionary_2018.xlsx"
NYC_311_DOCS = "data-dictionary.xlsx"

REQUESTS_CHUNK_SIZE = 4096
```

```python
if not os.path.isfile(NYC_311_RAW_DATA):
    temp_path = NYC_RAW_DATA + ".part"
    response = requests.get(NYC_311_URL, stream=True)
    response.raise_for_status()
    progress = tqdm.tqdm(
        # 12.3e9iB is the estimated size, just to get a rough idea of download time
        total=int(response.headers.get("content-length", 13.2e9)),
        unit="iB",
        unit_scale=True
    )

    with open(temp_path, "wb") as output_fp:
        for chunk in response.iter_content(REQUESTS_CHUNK_SIZE):
            progress.update(len(chunk))
            output_fp.write(chunk)

    progress.close()
    os.rename(temp_path, NYC_311_RAW_DATA)
```

```python
if not os.path.isfile(NYC_311_DOCS):
    response = requests.get(NYC_311_DOCS_URL)
    response.raise_for_status()
    with open(NYC_311_DOCS, "wb") as output_fp:
        output_fp.write(response.content)
```

# 2. Filter out fireworks incidents
\>12GB is far too much for a normal computer to be able to process without resorting to out-of-core processing methods. However, the number of 311 calls dealing with fireworks is likely to be a very small fraction of this, so we can do some filtering to make the data more manageable.

```python
# we could use a streaming csv or pandas solution, but using a compiled regular
# expression is likely to be far faster
RE_FIREWORKS = re.compile(r"fireworks\s*", re.IGNORECASE)
```

```python
if not os.path.isfile(NYC_311_TRIMMED_DATA):
    temp_path = NYC_311_TRIMMED_DATA + ".part"
    
    with open(NYC_311_RAW_DATA, "r") as input_fp,\
    open(temp_path, "w") as output_fp:
        # write out header
        output_fp.write(input_fp.readline())
        
        # filter out remaining rows
        # ~23.1M rows according to the dashboard, again to estimate time
        for line in tqdm.tqdm(
                input_fp,
                total=23.1e6,
                unit=" rows",
                unit_scale=True
            ):
            if RE_FIREWORKS.search(line):
                output_fp.write(line)
                
    os.rename(temp_path, NYC_311_TRIMMED_DATA)
```

# 3. Data cleanup, exploration, type conversions, etc

```python
df_311 = pandas.read_csv(NYC_311_TRIMMED_DATA)
df_311.sample(5)
```

```python
def sample_not_null(df: T_DataFrame, column: str) -> typing.Any:
    subset = df_311[column]
    if len(subset) > 0:
        return subset.sample(1).iloc[0]

functools.reduce(
    lambda left, right: pandas.merge(
        left, right, left_index=True, right_index=True
    ),
    [
        
        # official documentation
        pandas.read_excel(NYC_311_DOCS, sheet_name="Data Dictionary")\
            .set_index("Column Name")[["Description", "Expected Values"]],
        
        # data types
        df_311.dtypes.rename("dtype"),
        
        # coverage
        (df_311.isnull().sum() / len(df_311)).rename("percent_null"),
        
        # sample data
        pandas.Series(df_311.columns, index=df_311.columns)\
            .apply(lambda column: sample_not_null(df_311, column))\
            .rename("sample"),
    ]
).fillna("") # just to clean up the output
```

There are about 1.17% of reports without a geocoded lat-lon; is there a listed location that we can use to geocode?

```python
print(df_311["Incident Address"][df_311["Location"].isnull()].dropna().to_list())
```

It seems like the NYC geocoding team did a really good job; the non-geocoded locations don't seem like they can be geocoded very easily. We have plenty of data so we can just drop these for the analysis.

From the above, it seems that only a few columns will be relevant, so let's go ahead and subset them now, as well as add some spatial types:

```python
df_311 = df_311[[
    "Created Date", "Closed Date", # datetime variables
    "Incident Zip", "City", "Borough", # locations IDs, for standard joining
    "Longitude", "Latitude",  # WGS84 coords, for geospatial joining and viz
]]

# date conversions
for date_column in ["Created Date", "Closed Date"]:
    df_311[date_column] = pandas.to_datetime(df_311[date_column])

# convert zip codes to mutable integers
df_311["Incident Zip"] = df_311["Incident Zip"].astype("Int32")

# spatial types
df_311["geometry"] = df_311[["Longitude", "Latitude"]]\
    .apply(shapely.geometry.Point, axis=1)
gdf_311 = geopandas.GeoDataFrame(df_311)
gdf_311.crs = {"init": "epsg:4326"}

# transform to web mercator for more convenient plotting
gdf_311 = gdf_311.to_crs({"init": "epsg:3857"})
```

```python
gdf_311_recent = gdf_311[gdf_311["Created Date"] > pandas.to_datetime("2020-06-01")]
(figure, axis) = pyplot.subplots(figsize=(7, 7))
gdf_311_recent.plot(ax=axis, markersize=1, alpha=0.1)
axis.set_xticks([])
axis.set_yticks([])
axis.set_title(
    "Firework complaints since 2020-06-01"
    "\n({:.2f}% of all firework complaints since {})".format(
        len(gdf_311_recent) / len(gdf_311) * 100,
        gdf_311["Created Date"].min().strftime("%Y-%m-%d")
    )
)
contextily.add_basemap(ax=axis, source=contextily.providers.CartoDB.Positron)
pyplot.show()
```
